```python
def t_series(start,end):
    time_series=df[['Date','PM25']][(df['Date'] >= start) & (df['Date'] <= end)]
    x=time_series.Date
    y=time_series.PM25
    plt.plot(x,y)
    plt.xlabel('Date')
    plt.ylabel('PM2.5 Values')
    plt.title('PM2.5 Time Series')
    return plt.show();
```
